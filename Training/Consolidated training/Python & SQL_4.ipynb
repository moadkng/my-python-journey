```python
import pandas as pd
import sqlalchemy
import yfinance as yf
```

```python
wiki = "https://en.wikipedia.org/wiki/"
```

```python
pd.set_option("display.max_rows", None)
pd.set_option("display.max_columns", None)
pd.set_option("display.width", None)
pd.set_option("display.max_colwidth", None)
```

```python
tickersDJ30 = pd.read_html(wiki + "Dow_Jones_Industrial_Average")[1].Symbol.to_list()
```

```python
def get_price_data(tickers):
    price_data = []
    for ticker in tickers:
        price_data.append(yf.download(ticker, start = "2018-01-01").reset_index())
    return price_data
```

```python
tickersDJ30
```

```python
price_data = get_price_data(tickersDJ30)
```

```python
engine = sqlalchemy.create_engine("sqlite:///D30StockPrices.db")
```

```python
for ele, ticker  in zip(price_data, tickersDJ30):
        ele.to_sql(ticker, engine, index = False)
```

```python
AAPL = pd.read_sql("AAPL", engine)
```

```python
t1 = AAPL.index[AAPL["Date"] == "2019-12-31"].tolist()
```
