```python
import sqlite3
import sqlalchemy
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
engine = sqlalchemy.create_engine("sqlite:///MyDataBaseTest.db")
```

```python
df_univers_gc = pd.read_sql("universgc", engine)
```

```python
df_univers_gc
```

```python
tickers = df_univers_gc["ticker"].tolist()
```

```python
tkrs = [* range(len(tickers))]
i = 0
for tkr in tickers:
    ext = tkr.split(" ")
    tkrs[i] = ext[0]
    i += 1
```

```python
tkrs
```

```python
for ele in tkrs:
    df_res = yf.download(ele, start = "2021-01-01")
    df_res.reset_index(drop = False, inplace = True)
    df_res.to_sql(ele, engine, if_exists = "append", index = False)
```
